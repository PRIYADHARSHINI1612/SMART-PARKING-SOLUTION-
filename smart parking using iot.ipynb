```python
# Import required libraries
import random
import time
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import clear_output

# Configuration
total_slots = 10  # You can change this for a bigger lot
parking_slots = ["Free"] * total_slots

# Simulate sensor values (1 = Occupied, 0 = Free)
def simulate_sensor_input():
    return [random.choice(["Occupied", "Free"]) for _ in range(total_slots)]

# Visualize Parking Slot Status
def display_parking_lot(status_list):
    clear_output(wait=True)
    print("Smart Parking Slot Status\n")
    for i, status in enumerate(status_list, 1):
        print(f"Slot {i}: {status}")
    print("\nOccupied:", status_list.count("Occupied"))
    print("Free     :", status_list.count("Free"))

# Run simulation for 5 cycles
for i in range(5):
    parking_slots = simulate_sensor_input()
    display_parking_lot(parking_slots)
    time.sleep(2)
```

```python
# Pie chart of current status
labels = 'Occupied', 'Free'
sizes = [parking_slots.count("Occupied"), parking_slots.count("Free")]
colors = ['red', 'green']
explode = (0.1, 0)# explode occupied slot

plt.figure(figsize=(5,5))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("Smart Parking Lot - Current Status")
plt.axis('equal')
plt.show()
```
